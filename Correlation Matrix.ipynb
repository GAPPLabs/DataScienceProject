# Correlation Matrix

Esta matriz de correlación tiene como objetivo el poder determinar las columnas (variables, en Ciencias de Datos) que poseen una relación de conocimiento entre los datos.

Para esto, se define lo siguiente:

- **Conjunto de datos.** Se elegió el conjunto de datos "futbol" que contiene información relacionada de diversos jugadores de distintos clubes de todo el mundo; esto con su información estadistica como jugador y su algunos datos personales.


- **Tipo de relación.** Un valor cercano a 1 significa relación proporcional y a -1 significa una relación inversa.

## Importación de librerias

Se importando todas las librerias necesarias para este proyecto, donde las mas importante es **pandas** ya que permite la lectura, manipulación y almacenamiento de los datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Lectura del conjunto de datos

Se lee todo el conjunto de datos elegido.

```python
dataframe = pd.read_csv("futbol.csv", parse_dates= True)
```

Posteriormente, se visualiza el dataframe (estructura que posee un conjunto de datos) leido.

```python
dataframe
```

## Visualización de los datos

Por ultimo, se visualiza la gráfica de tipo correlación para observar las relaciones mas significativas entre las columnas del dataframe leido.

```python
plt.figure(figsize=(100,80))
sns.heatmap(dataframe.corr(), cmap ='Blues', linewidths = 0.30, annot = True)
```
